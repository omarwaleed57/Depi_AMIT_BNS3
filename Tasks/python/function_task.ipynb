```python
# Question 2

import math
def prime_nums(max: int):
    '''
    This function gets prime numbers to the maximum given number

    par_1: the maximum number from the user
    type_par_1: int

    return: prime numbers
    type_return: list
    '''

    primes = []
    
    for i in range(2,max+1):
        divisors = []
        j = 2
        while j <= int(math.sqrt(i)):       # Don't need more than the square root cause the lower numbers would have given the bigger ones as a reminder
            if (i) % (j) == 0 :
                divisors.append(j)
            j += 1
                
        if len(divisors) == 0:
            primes.append(i)

    return primes

# prime_nums(1000)


def twin_primes(max: int):
    '''
    This function gets prime numbers from another function to the max number given and gives out the twin primes

    par_1: the maximum number to check for
    type_par_1: int

    return: prints the twin primes
    type_return: str
    '''
    primes = prime_nums(max)
    primes.remove(2)
    for i in range(len(primes)-1):       # Don't need the last number cause no number after it in the list
        if primes[i+1] - primes[i] == 2:
            print(f'{primes[i]} and {primes[i+1]}')


# Change the number here or try the function in a new cell
twin_primes(100)
```

```python
#  Question 3

def prime_factors(num: int):
    '''
    This function gives the prime factors of a number

    par_1: user given number
    type_par_1: int

    return: returns the list of prime factors of the given number
    return_type: list
    '''
    factors = []
    while num > 1:
        i = 2
        while i <= num:
            if num % i == 0:
                factors.append(i)
                num = num/i
            else:
                i += 1
    return factors
prime_factors(56)
```

```python
#  Question 4

import math
def DecToBin(num: int):
    '''
    This function converts positive dicimal numbers into binary numbers
    
    par_1: takes the decimal number to be converted
    type_par_1: positive int

    return: returns the equivalent binary number
    type_return: int
    '''
    if num <= 0:
        print('Function only converts positive integers')
        return
    rem = num
    Binary = []
    while rem > 0:
        if rem % 2 == 1:
            Binary.insert(0,1)
        else:
            Binary.insert(0,0)
        rem = int(rem/2)
    binnum = 0
    for i in range(len(Binary)):
        binnum += (10**(i))*Binary[i]
    return binnum

DecToBin(11)
```
