```python
import requests
from bs4 import BeautifulSoup

html_text = requests.get(
    'https://wuzzuf.net/search/jobs?a=spbg&q=machine%20learning')

soup = BeautifulSoup(html_text.text, 'lxml')
jobs = soup.find_all('div', class_='css-pkv5jc')

for job in jobs:
    company_name = job.find('a', class_='css-17s97q8')
    keywords = job.find_all('a', class_='css-o171kl') + job.find_all('a', class_ = 'css-5x9pm1')
    for keyword in range(len(keywords)):
        keywords[keyword] = keywords[keyword].text
    published_date = job.find('div', class_='css-4c4ojb')
    Location = job.find('span', class_ = 'css-5wys0k')

    if company_name and keywords and published_date and Location:
        company_name = company_name.text.strip()
        Location = Location.text.strip()
        keywords = ', '.join(keywords)
        published_date = published_date.text.strip()

        print(f'''   
        Company Name     : {company_name}\b\b
        Location         : {Location}
        Published Date   : {published_date}
        keywords         : {keywords}

    =======================================
        ''')
```
